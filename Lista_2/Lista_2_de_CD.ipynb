```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("diabetes_DS.csv")

df
```

```python
X = df[df.columns[:-1]].values
y = df['Outcome'].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42, stratify=y)
```



### A) Complete a variável 'K' com valores 'k' condizentes para o KNN.

### B) Adicione, nas listas de treino e teste, as acurácias do modelo knn treinado. Vide documentação para obter a acurácia.

* Obs: utilizar X_train, y_train para acurácia do treino e X_test e y_test para acurácia do teste.


Para mais informações:
[sklearn.neighbors.KNeighborsClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html)

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

K =  range(1, 100)
accuracy_training = []
accuracy_testing  = []

for i in K:
  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(X_train, y_train)

  y_train_pred = knn.predict(X_train)
  y_test_pred = knn.predict(X_test)

  treino_acauracia = accuracy_score(y_train, y_train_pred)
  teste_acuracia = accuracy_score(y_test, y_test_pred)

  accuracy_training.append(treino_acauracia)
  accuracy_testing.append(teste_acuracia)
```

### C) Gere um plot com a acurácia dos dados de treinamento e outro com os dados de teste. Os eixos devem ser compostos da maneira a seguir:
* Eixo x: valores de k
* Eixo y: valores da acurácia

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.title('Acurácia variando com os valores de K')

plt.plot(K, accuracy_training, label='Treinamento', marker='o')
plt.plot(K, accuracy_testing, label='Teste', marker='o')

plt.xlabel('Valores de K')
plt.ylabel('Acurácia')
plt.title('Acurácia nos Dados de Treinamento e Teste')
plt.legend()

plt.show()
```

## A) Explique como o *fitting* da Regressão Logística se diferencia da Regressão Linear.

RESP: A regressão linear busca correlações entre duas variáveis numéricas, a regressão pode ter variaveis numéricas ou categóricas.

## B) Em quais contextos a Regressão Logística é preferível de se utilizar do que a Linear?

RESP: Ela pode ser utilizada quando o problema puder ser categorizado, por exempo: "Qual a probabilidade chances de um funcionário ser demitido?". Algo categórico, ou ele é demitido ou não.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import accuracy_score

df = pd.read_csv("diabetes_DS.csv")
df
```

```python
X = df[df.columns[:-1]].values
y = df['Outcome'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42, stratify=y)
```

```python
# Regressão Linear
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
linear_pred = linear_reg.predict(X_test)
linear_pred = [1 if pred >= 0.5 else 0 for pred in linear_pred]

# Acurácia da Regressão Linear
linear_accuracy = accuracy_score(y_test, linear_pred)
print("Acurácia - Regressão Linear:", linear_accuracy)

# Regressão Logística
logistic_reg = LogisticRegression(solver='liblinear')
logistic_reg.fit(X_train, y_train)
logistic_pred = logistic_reg.predict(X_test)

# Acurácia da Regressão Logística
logistic_accuracy = accuracy_score(y_test, logistic_pred)
print("Acurácia - Regressão Logística:", logistic_accuracy)
```

```python
from pymfe.mfe import MFE
import pandas as pd
import numpy as np
from scipy.spatial.distance import euclidean

data1 = pd.read_csv('IrisNormalizad.csv')
data2 = pd.read_csv('WineNormalizad.csv')

mfe1 = MFE() # Cria uma nova instância da classe para extrair as características da Base de Dados
mfe1.fit(data1.values) # Ajusta o estrator de caracteristicas com os valores do Data frame
_, vetor_caracteristica1 = mfe1.extract() # Extrai os Vetores de Caraceterísricas da Base de Dados
vetor_caracteristica1 = np.array(vetor_caracteristica1) # Armazena os valores

mfe2 = MFE()
mfe2.fit(data2.values)
_, vetor_caracteristica2 = mfe2.extract()
vetor_caracteristica2 = np.array(vetor_caracteristica2)

vetor_caracteristica1[np.logical_not(np.isfinite(vetor_caracteristica1))] = np.nanmean(vetor_caracteristica1) #Trata os valores não finitos e substitiui pela média dos valores finitos
vetor_caracteristica2[np.logical_not(np.isfinite(vetor_caracteristica2))] = np.nanmean(vetor_caracteristica2)

distancia_euclidiana = euclidean(vetor_caracteristica1, vetor_caracteristica2)

print(f"Distância Euclidiana das Bases de Dados: {distancia_euclidiana:.4f}")
```
